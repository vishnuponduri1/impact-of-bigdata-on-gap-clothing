```python
import urllib.request, urllib.error, urllib.parse
import json
import csv
import html
import time
from bs4 import BeautifulSoup
import random
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.service import Service
from webdriver_manager.firefox import GeckoDriverManager
firefox_options = Options()
#firefox_options.add_argument('--headless')
firefox_options.add_argument('--no-sandbox')
firefox_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()), options = firefox_options)
```

```python
url='https://oldnavy.gap.com/browse/category.do?cid=1035712&#department=136&facetOrder=department:136'
driver.get(url)
time.sleep(3)
html=driver.page_source
html
```

```python
soup=BeautifulSoup(html,'html.parser')
```

```python
with open('old_navy_color_data.csv','w',encoding="utf-8",newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Parameter","Parameter Value","Data Value", "Total Items"])
    total_items=soup.find('div',class_="category-page-p6sjqa").text[:-6]
    color_set=soup.find_all('div',class_='facetbar-item-colorswatch category-page-5h7d9x')
    for color in color_set:
        color_details=color.find('input',class_="category-page-40vz4u")
        color_name=color_details['value']
        color_id=color_details['id']
        url='https://oldnavy.gap.com/browse/category.do?cid=1035712&#department=136&color=' + color_id + '&facetOrder=department:136,color:1005'
        driver.get(url)
        time.sleep(3)
        html2=driver.page_source
        soup2=BeautifulSoup(html2,'html.parser')
        item_count=soup2.find('div', class_="category-page-p6sjqa").text[:-5].strip()
        writer.writerow(["Color",color_name,item_count,total_items])
```
